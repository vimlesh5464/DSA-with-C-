#String Definition
1.What is a string in programming?
A string is a sequence of characters used to store text.
In C++, you can use:
char[] (character array) → C-style string
std::string → C++ string class with built-in methods

2.Difference between character array and string class/object|
Feature          | `char[]` (C-style)    | `std::string` (C++ class)         |
| ---------------- | --------------------- | --------------------------------- |
| Storage          | Contiguous char array | Managed internally, dynamic       |
| Null-termination | Required (`\0`)       | Not required                      |
| Methods          | Manual operations     | `.length()`, `.substr()`, etc.    |
| Mutability       | Mutable               | Mutable in C++, Immutable in Java |
                              |


#Memory Representation
1.How strings are stored in memory (contiguous array of characters)

In C++ char str[] = "Hello";, memory looks like:

['H']['e']['l']['l']['o']['\0']
std::string internally uses a dynamic array to store characters; may include capacity greater than current length for efficiency.


2.Null-terminated strings (\0) in C vs string objects in C++/Java
C (char[]) → last character is '\0' to mark end
C++ (std::string) → no need for \0, length tracked internally
Java/Python strings → immutable objects, length tracked internally

#String Length
Difference between strlen() in C and .length() / .size() in C++/Java
strlen(char* str) → counts characters until \0, O(n)
std::string::length() or std::string::size() → O(1), because length is stored as a member variable

#String Immutability
Strings in C (mutable) vs Java/Python (immutable)
C/C++ char[] or std::string → mutable
Java/Python strings → immutable
How immutability affects operations like concatenation
Concatenation creates a new string, old string unchanged

Example:

std::string s1 = "Hello";
std::string s2 = s1 + " World"; // creates a new string

```python
#Character Encoding
ASCII vs Unicode (UTF-8/UTF-16)
ASCII → 7-bit, supports English letters and basic symbols
UTF-8 → variable length (1–4 bytes), supports all Unicode characters
UTF-16 → 2 or 4 bytes, supports Unicode


Why encoding matters for string comparison and length
In ASCII: 1 char = 1 byte → std::string.size() = number of characters
In UTF-8: some characters >1 byte → size in bytes ≠ number of logical characters
Must use proper encoding for correct comparison and length
```

```python
2️⃣ String Operations
Basic Operations
Concatenation → s1 + s2 or s1.append(s2)
Copy → std::string s2 = s1;
Compare → s1 == s2, s1.compare(s2)
Substring → s1.substr(start, length)
Indexing / Slicing / Finding
s1[i] → access character at index i
s1.find("abc") → find substring
```

```python
Searching & Matching
How strstr / std::string::find works internally
Naive approach: check every substring → O(n*m)
Optimized: KMP, Rabin-Karp for large strings → O(n + m)
Exact match / prefix match / suffix match
Exact match: full string equality
Prefix: starts with some substring → s.compare(0, len, prefix)
Suffix: ends with substring → s.compare(s.size()-len, len, suffix)
```

```python
Sorting & Comparison
Lexicographical ordering
Compare characters from left to right
'a' < 'b' < 'c' < ...
Case-sensitive vs case-insensitive comparison
Case-sensitive: 'A' < 'a'
Case-insensitive: convert both to lower/upper before compare
```

#Common Functions (C / C++ / Java)
| Operation         | C (`char*`)         | C++ (`std::string`)         | Java (`String`)                      |
| ----------------- | ------------------- | --------------------------- | ------------------------------------ |
| Length            | `strlen(s)`         | `s.size()` / `s.length()`   | `s.length()`                         |
| Compare           | `strcmp(s1,s2)`     | `s1.compare(s2)`            | `s1.equals(s2)` / `s1.compareTo(s2)` |
| Copy / Assignment | `strcpy(dest, src)` | `s2 = s1`                   | `s2 = s1`                            |
| Concatenate       | `strcat(s1,s2)`     | `s1 + s2` / `s1.append(s2)` | `s1 + s2` / `s1.concat(s2)`          |
| Substring / Find  | N/A                 | `s.substr()` / `s.find()`   | `s.substring()` / `s.indexOf()`      |
| Replace           | N/A                 | `s.replace(pos, len, str)`  | `s.replace()`                        |


```python
3️⃣ Advanced Concepts
String Algorithms
Naive substring search → check all substrings → O(n*m)
KMP → preprocess pattern → O(n+m) search
Rabin-Karp → rolling hash → O(n+m) average
Trie / Prefix tree → store multiple strings efficiently → fast prefix queries
Suffix array / Suffix tree → fast substring search, longest repeated substring
```

```python
Pattern Matching
Wildcards (* / ?)
* → matches 0 or more characters
? → matches exactly 1 character
Regular expressions
Used for complex matching (e.g., emails, phone numbers)
```

```python
Hashing:
Rolling hash
Efficient substring comparison
Used in Rabin-Karp, plagiarism detection
Applications
Check anagrams, substring search, repeated patterns
```

```python
String Internals:
1. Immutable strings:
Stored in memory pool in Java
C++ strings are mutable, but optimization possible with SSO (Small String Optimization)
Copy-on-write
Avoids unnecessary copying until modification
```
